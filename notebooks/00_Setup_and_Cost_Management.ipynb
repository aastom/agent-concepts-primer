## 1. Setting Up API Keys

Our agent will need to connect to services like OpenAI (for the LLM) and Tavily (for web search). These services use API keys to authenticate requests.

**NEVER hardcode your API keys in your notebooks or scripts.** This is a major security risk. If you accidentally share your code, anyone can use your key and run up a large bill.

The best practice is to use a `.env` file.

1.  **Find `.env.example`**: In the root directory of this project, you will find a file named `.env.example`.

2.  **Rename to `.env`**: Create a copy of this file and name it `.env`. The `.gitignore` file is configured to prevent this file from ever being committed to git.

3.  **Get Your Keys**:
    *   **OpenAI**: Go to [https://platform.openai.com/api-keys](https://platform.openai.com/api-keys), create a new secret key, and copy it.
    *   **Tavily**: Go to [https://app.tavily.com/home](https://app.tavily.com/home), sign up for an account, and copy your API key. The free tier is sufficient for this module.

4.  **Edit the `.env` file**: Open your new `.env` file and paste your keys in the appropriate places. It should look like this:

```
OPENAI_API_KEY="sk-xxxxxxxxxxxxxxxxxxxxxxxxxxxxxx"
TAVILY_API_KEY="tvly-xxxxxxxxxxxxxxxxxxxxxxxxxxxxxx"
```

## 2. Understanding Model Choices & Cost

The cost of running an agent is directly tied to the LLM you choose. Different models have different prices, speeds, and capabilities.

- **High-Capability Models (e.g., `gpt-4o`)**: Very smart, great at reasoning, but more expensive and slower.
- **Balanced Models (e.g., `gpt-3.5-turbo`)**: A good mix of performance and cost. **This is the default for our learning module.**
- **Local Models (e.g., via Ollama)**: If you have a powerful local machine, you can run models for free. This is a great option for development but requires more setup.

For this primer, all examples will default to `gpt-3.5-turbo` to keep costs as low as possible while still providing a good learning experience. We will explicitly show you where the model is defined so you can change it if you wish.

## 3. Loading API Keys into the Environment

Now, let's see how we can load the keys from our `.env` file so that Python libraries like `langchain` and `openai` can use them automatically.

We use the `python-dotenv` library, which we installed via `requirements.txt`.

````python
from dotenv import load_dotenv, find_dotenv
import os

# `load_dotenv` will look for a .env file in the project directory
# `find_dotenv` helps to find the .env file in the parent directories if the script is run from a subdirectory
success = load_dotenv(find_dotenv())

if not success:
    print("Could not find .env file. Please create one following the instructions.")
else:
    print("Found and loaded .env file.")

# Now, let's check if the OpenAI API key is loaded
openai_api_key = os.getenv("OPENAI_API_KEY")

if openai_api_key:
    # We print only the first 5 and last 4 characters to avoid exposing the key
    print(f"OpenAI API Key loaded successfully: {openai_api_key[:5]}...{openai_api_key[-4:]}")
else:
    print("OpenAI API Key not found. Make sure it is set in your .env file.")
````

### How it Works

When you call `load_dotenv()`, the library reads your `.env` file and adds its contents to the environment variables of the current session. 

Libraries like `langchain_openai` are designed to automatically look for environment variables like `OPENAI_API_KEY`. If the variable is present, they will use it to authenticate. This means you only need to run the `load_dotenv()` command at the beginning of your script or notebook, and everything else works automatically. You don't need to pass your key around in your code.
